```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
%matplotlib inline
```

```python
stocks =["INFY", "AAPL", "GOOGL", "TSLA", "BTC-USD", "005930.KS", "DOGE-USD", "RELIANCE.NS", "META", "IBM", "ORCL", "ERF.PA", "NVDA", "SOL-USD"]

```

```python
import datetime
# dict={}
# for s in stocks:
#    da=yf.Ticker(s)
#    da=da.history(period="1y")
#    da=da.reset_index()
#    dict[s]=da
yesterday = datetime.datetime.today() - datetime.timedelta(days=1)
yesterday_str = yesterday.strftime("%Y-%m-%d")

# Get data from 1 year ago to yesterday
one_year_ago = (datetime.datetime.today() - datetime.timedelta(days=365)).strftime("%Y-%m-%d")

for s in stocks:
    da = yf.Ticker(s)
    da = da.history(start=one_year_ago, end=yesterday_str)  # Corrected: Use start & end instead of period
    da = da.reset_index()
    dict[s] = da
```

```python
df=[]
for stock,data in dict.items():
    data["Stock"]=stock
    df.append(data)
    comb_df=pd.concat(df,ignore_index=True)
```
