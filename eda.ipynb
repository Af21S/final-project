```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
```

```python
stocks = [
    'Infy', 'AAPL', 'Goog', 'Tsla', 'Doge-usd',  'Meta', 'IBM', 'Orcl', 'Esf0.ha', 'Nvda', 'Nflx', 'Sol-usd'
]
```

```python
dict={}
for s in stocks:
   da=yf.Ticker(s)
   da=da.history(period="1y")
   da=da.reset_index()
   dict[s]=da
   
```

```python
print(yf.Ticker(s))
```

```python
dict["Infy"]
```

```python
df=[]
for stock,data in dict.items():
    data["Stock"]=stock
    df.append(data)
    comb_df=pd.concat(df,ignore_index=True)
```
